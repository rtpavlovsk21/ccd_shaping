```python
import cv2;
import numpy as np;
from matplotlib.colors import LogNorm;
from matplotlib import pyplot as plt;
from matplotlib.widgets import Slider;
import h5py;


def coherent_interference(image):
    import ccd_plotter as ccp;
    import ccd_tools as cct;

    fimage=cv2.dft(image.astype(np.float32),flags=cv2.DFT_COMPLEX_OUTPUT);
    
    k=10;
    i=np.zeros((image.shape[0],image.shape[1]));
    i[int(image.shape[0]/2)+1,int(image.shape[1]/2)+1]=1;
    g=cv2.GaussianBlur(i.astype(np.float32),(image.shape[0],image.shape[1]),0.5);
    fg=cv2.dft(g,flags=cv2.DFT_COMPLEX_OUTPUT);
    fg2=(fg**2).sum(-1);
    print fg2.shape;
    ifg=fg/fg2[:,:,None];
    res=cv2.mulSpectrums(fimage,ifg,0);
    res=cv2.idft(res,flags=cv2.DFT_SCALE|cv2.DFT_REAL_OUTPUT);
    res=np.roll(res, -res.shape[0]//2,0);
    res=np.roll(res, -res.shape[1]//2,1);
    res=res#-image;
    no_neg=np.less(res,0);
    res[no_neg]=0;
    plt.imshow(res,'cubehelix',interpolation='none');
    plt.colorbar();
    
    plt.figure();
    
    plt.imshow(image,'cubehelix',interpolation='none');
    plt.colorbar();
    
    plt.show();
    
    #fimage=cv2.idft(fimage*fg);
    #plt.imshow(fimage,'cubehelix');
    #plt.colorbar();
    #plt.figure();
    #plt.imshow(image,'cubehelix');
    #plt.colorbar();
    #plt.show();

def weiner_attempt(image):
    fimage=cv2.dft(image.astype(np.float32),flags=cv2.DFT_COMPLEX_OUTPUT);
    
    k=10;
    i=np.zeros((image.shape[0],image.shape[1]));
    i[int(image.shape[0]/2)+1,int(image.shape[1]/2)+1]=1;
    g=cv2.GaussianBlur(i.astype(np.float32),(image.shape[0],image.shape[1]),0.5);
    fg=cv2.dft(g,flags=cv2.DFT_COMPLEX_OUTPUT);
    fg2=(fg**2).sum(-1);
    
    ifg=fg/fg2[:,:,None];
    print fg2.shape
    print np.mean(fg2);
    
    ifg=(ifg)/(np.mean(fg2)*5+ifg*ifg)
    res=cv2.mulSpectrums(fimage,ifg,0);
    res=cv2.idft(res,flags=cv2.DFT_SCALE|cv2.DFT_REAL_OUTPUT);
    res=np.roll(res, -res.shape[0]//2,0);
    res=np.roll(res, -res.shape[1]//2,1);
    res=res#-image;
    no_neg=np.less(res,0);
    res[no_neg]=0;
    plt.imshow(res,'cubehelix',interpolation='none');
    plt.colorbar();
    
    plt.figure();
    
    plt.imshow(image,'cubehelix',interpolation='none');
    plt.colorbar();

    plt.figure();
    
def butterworth(D=1,dims=0,n=1):
    h=np.zeros(dims);
    centers=np.asarray([dims[0]//2.+1,dims[1]//2.+1]);
    print centers;
    array_j=np.asarray( [x for x in range(0,dims[1])] );
    for k in range(0,dims[0]):
        d = array_j - centers[1];
        d = np.sqrt( d**2 + (k-centers[0])**2 );
        h[k,:]=1/(1+(np.sqrt(2)-1)*(D/d)**(2*n));
    return h;

def butterworth_filter(image,D=10):
    abfg=np.abs(np.fft.fft2(image)) ;

    but=np.fft.fftshift(butterworth(D=D,dims=image.shape));
    aabfg=np.fft.fft2(image);
    aabfg=aabfg*but;

    aabfg=np.fft.ifft2(aabfg);
    return aabfg;

def imshow_mag(image,ax=np.asarray([]),fig=[],cb=[]):
    aabfg=np.fft.fft2(image.astype(np.float));
    colorbar=False;
    if(not ax.shape[0]):
        fig,ax=plt.subplots(3);
        colorbar=True;
        cb=[];

    aabfg=1+np.abs(aabfg);
    aabfg=np.fft.fftshift(aabfg);
    cs=ax[0].imshow(aabfg,'cubehelix',interpolation='none',norm=LogNorm(vmin=np.max([(aabfg).min()]),vmax=(aabfg).max()) );
    if(colorbar):
        cb1=fig.colorbar(cs,ax=ax[0]);
        cb.append(cb1);
    cb[0].set_clim(vmin=np.max([aabfg.min(),1E-3]), vmax=aabfg.max());    

    cs=ax[1].imshow(2+np.abs(image),'cubehelix',interpolation='none',norm=LogNorm(vmin=np.max([1E-9,np.abs(image).min()]),vmax=np.abs(image).max()) );
    if(colorbar):
        cb2=fig.colorbar(cs,ax=ax[1]);
        cb.append(cb2);
    cb[1].set_clim(vmin=np.max([1E-9,np.abs(image).min()]), vmax=np.abs(image).max());

    ax[2].cla();
    ax[2].hist(image.ravel(),5000,histtype='stepfilled');
    ax[2].set_yscale('log');
    return fig,ax,cb;

def expand_bounding_box(bb,border=0):
    bb=np.asarray(bb);
    bb[2:]+=bb[:2];

    expand=np.asarray([-border,-border,border,border]);
    bb+=expand;
    negative= np.less(bb,0);
    bb[negative]=0;
    return bb;


def segment(image_array):        
    etrack_bb=[];
    nframes=image_array.shape[2];
    bw_image=(image_array>200).astype(np.uint8);

    kernel_lst=[np.ones((3,3),dtype=np.uint8)];
    hist_lst=[];

    kcounter=0;
    for kernel in kernel_lst:
        evt_lst=[];
        nevents=0;
        for frame in range(0,nframes):
            frame_bb=[];
            tbw_image=bw_image[:,:,frame];
            kernel_erode=np.ones((3,3),dtype=np.uint8);
            tbw_image=cv2.erode(tbw_image,kernel_erode,iterations=1);
            tbw_image=cv2.dilate(tbw_image,kernel,iterations=2);
        
            nmarkers,markers=cv2.connectedComponents(tbw_image);
            #marker 0 is probably background
            events=np.zeros(nmarkers-1);
            nevents+=nmarkers-1;
            for k in range(1,nmarkers):
                events[k-1]=np.sum( image_array[markers==k,frame] );
                event_bb=cv2.boundingRect((markers==k).astype(np.uint8));
                event_bb=expand_bounding_box(event_bb,border=0);
                frame_bb.append(event_bb);
            evt_lst.append(events);
            etrack_bb.append(frame_bb);
        
        events=np.zeros(nevents);
        for evt in evt_lst:
            events[nevents-evt.shape[0]:nevents]=evt;
            nevents-=evt.shape[0];
    return np.asarray(evt_lst),np.asarray(etrack_bb);

#fil=h5py.File('data/ccd_images.h5','r');
    
#images=fil['ba133'][:,:,:29];
#med_image=np.median(images,axis=2);
#images-=med_image[:,:,None];
fil=h5py.File('data/short.h5','r');
images=fil['ba133'];
med_image=np.median(images,axis=2);
image=np.rot90(med_image);
bo=np.greater(image,0);
if(bo.any()):
    print bo;
fimage=butterworth_filter(image,D=10);
fig,ax,cb=imshow_mag(fimage);

axcolor='lightgoldenrodyellow';
axfreq=plt.axes([0.25,0.05,0.65,0.02],axisbg=axcolor);

dfreq = Slider(axfreq, 'Freq', 0.1, 7, valinit=6);

def update(val):
    d=10**dfreq.val;
    fimage=butterworth_filter(image,D=d);
    imshow_mag(fimage,fig=fig,ax=ax,cb=cb);
    fig.canvas.draw_idle();

dfreq.on_changed(update);
plt.show();
    
#output=np.less(image,80);
#image[output]=1;
#ccp.plot_ccd(image);
    
#cct.make_histogram(image_array=images,dims=image.shape);
    
```
